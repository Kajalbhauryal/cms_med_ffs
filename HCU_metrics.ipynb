```python
df_hcu_state = df_hcu_utilization[(df_hcu_utilization['BENE_GEO_LVL'] == 'State')]

df_hcu_state = df_hcu_state.replace('*', pd.NA)

df_hcu_state = df_hcu_state[~df_hcu_state['BENE_GEO_DESC'].isin(['Territory', 'ZZ'])]

er_visits_clean = df_hcu_state[['YEAR', 'BENE_GEO_DESC', 'BENE_GEO_LVL', 'ER_VISITS_PER_1000_BENES']]
```

```python
er_visits_clean['ER_VISITS_PER_1000_BENES'] = pd.to_numeric(er_visits_clean['ER_VISITS_PER_1000_BENES'], errors='coerce')

state_avg_er_visits = er_visits_clean.groupby('BENE_GEO_DESC')['ER_VISITS_PER_1000_BENES'].mean().reset_index()

state_avg_er_visits = state_avg_er_visits.sort_values(by='ER_VISITS_PER_1000_BENES', ascending=False)

state_avg_er_visits
```

```python
import matplotlib.pyplot as plt
import pandas as pd

state_er_visits = (
    er_visits_clean[er_visits_clean['BENE_GEO_LVL'] == 'State']
    .groupby(['BENE_GEO_DESC', 'YEAR'], as_index=False)
    .agg({'ER_VISITS_PER_1000_BENES': 'mean'})
)

top_states = (
    state_er_visits
    .groupby('BENE_GEO_DESC')['ER_VISITS_PER_1000_BENES']
    .mean()
    .sort_values(ascending=False)
    .head(5)
    .index
    .tolist()
)

color_palette = ['red', 'blue', 'green', 'orange', 'purple']
state_color_map = dict(zip(top_states, color_palette))

national_avg_er = er_visits_clean.groupby('YEAR')['ER_VISITS_PER_1000_BENES'].mean().reset_index()


plt.figure(figsize=(14, 8))

for state in state_er_visits['BENE_GEO_DESC'].unique():
    state_data = state_er_visits[state_er_visits['BENE_GEO_DESC'] == state]
    plt.plot(
        state_data['YEAR'], 
        state_data['ER_VISITS_PER_1000_BENES'], 
        color='lightgray', 
        linewidth=0.7, 
        alpha=0.4
    )

for state in top_states:
    state_data = state_er_visits[state_er_visits['BENE_GEO_DESC'] == state]
    plt.plot(
        state_data['YEAR'], 
        state_data['ER_VISITS_PER_1000_BENES'], 
        color=state_color_map[state], 
        linewidth=2.5, 
        label=state
    )

plt.plot(
    national_avg_er['YEAR'], 
    national_avg_er['ER_VISITS_PER_1000_BENES'], 
    color='black', 
    linewidth=3, 
    linestyle='--', 
    label='National Average'
)

plt.xlabel('Year')
plt.ylabel('Avg ER Visits per 1,000 Medicare FFS Beneficiaries')
plt.title('Medicare ER Visits Trends (2014–2023) - Top 5 States by Average')
plt.grid(True)
plt.legend(title='State')
plt.tight_layout()
plt.savefig('er_visits_trends_top5_states_avg.png', dpi=300)
plt.ylim(350, 1200)
plt.yticks(range(350, 1200, 100))
plt.show()
```
