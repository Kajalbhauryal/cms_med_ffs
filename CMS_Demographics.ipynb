```python
import pandas as pd
```

```python
df_cms = pd.read_csv('2014-2023 Medicare Fee-for-Service Geographic Variation Public Use File.csv')
```

```python
df_cms_clean = df_cms.replace('*', pd.NA)
```

```python
df_demographics = df_cms_clean[['YEAR', 'BENE_GEO_LVL', 'BENE_GEO_DESC', 'BENE_AGE_LVL', 'BENES_TOTAL_CNT', 'BENES_WTH_PTAPTB_CNT', 'BENES_FFS_CNT', 'BENES_MA_CNT', 'MA_PRTCPTN_RATE', 'BENE_AVG_AGE', 'BENE_FEML_PCT', 'BENE_MALE_PCT', 'BENE_RACE_WHT_PCT', 'BENE_RACE_BLACK_PCT', 'BENE_RACE_HSPNC_PCT', 'BENE_RACE_OTHR_PCT', 'BENE_DUAL_PCT', 'BENE_AVG_RISK_SCRE']]
```

```python
df_demographics.loc[:, 'BENE_FEML_PCT'] = pd.to_numeric(df_demographics['BENE_FEML_PCT'], errors='coerce')
df_demographics.loc[:, 'BENE_MALE_PCT'] = pd.to_numeric(df_demographics['BENE_MALE_PCT'], errors='coerce')
df_demographics.loc[:, 'BENE_RACE_WHT_PCT'] = pd.to_numeric(df_demographics['BENE_RACE_WHT_PCT'], errors='coerce')
df_demographics.loc[:, 'BENE_RACE_BLACK_PCT'] = pd.to_numeric(df_demographics['BENE_RACE_BLACK_PCT'], errors='coerce')
df_demographics.loc[:, 'BENE_RACE_HSPNC_PCT'] = pd.to_numeric(df_demographics['BENE_RACE_HSPNC_PCT'], errors='coerce')
df_demographics.loc[:, 'BENE_RACE_OTHR_PCT'] = pd.to_numeric(df_demographics['BENE_RACE_OTHR_PCT'], errors='coerce')
df_demographics.loc[:, 'BENE_DUAL_PCT'] = pd.to_numeric(df_demographics['BENE_DUAL_PCT'], errors='coerce')
df_demographics.loc[:, 'BENE_AVG_RISK_SCRE'] = pd.to_numeric(df_demographics['BENE_AVG_RISK_SCRE'], errors='coerce')
```

```python
numeric_columns = [
    'BENES_TOTAL_CNT', 'BENES_WTH_PTAPTB_CNT', 'BENES_FFS_CNT', 'BENES_MA_CNT', 
    'MA_PRTCPTN_RATE', 'BENE_AVG_AGE', 'BENE_FEML_PCT', 'BENE_MALE_PCT',
    'BENE_RACE_WHT_PCT', 'BENE_RACE_BLACK_PCT', 'BENE_RACE_HSPNC_PCT', 
    'BENE_RACE_OTHR_PCT', 'BENE_DUAL_PCT', 'BENE_AVG_RISK_SCRE'
]
for column in numeric_columns:
    df_demographics_clean[column] = pd.to_numeric(df_demographics_clean[column], errors='coerce')
non_numeric_columns = df_demographics_clean.select_dtypes(exclude=['number']).columns
print("Non-numeric columns after conversion:", non_numeric_columns)
```

```python
df_demographics_clean.rename(columns={
    'BENE_GEO_DESC': 'State/County',  
    'BENE_GEO_LVL': 'Geographical Level',
    'BENE_AGE_LVL': 'Age Level',
    'BENE_FEML_PCT': 'Female Percentage',
    'BENE_MALE_PCT': 'Male Percentage',
    'BENE_RACE_WHT_PCT': 'White Percentage',
    'BENE_RACE_BLACK_PCT': 'Black Percentage',
    'BENE_RACE_HSPNC_PCT': 'Hispanic Percentage',
    'BENE_RACE_OTHR_PCT': 'Other Race Percentage',
    'BENE_DUAL_PCT': 'Dual Eligibility Percentage',
    'BENE_AVG_RISK_SCRE': 'Average HCC Risk Score',
    'BENES_TOTAL_CNT': 'Total Beneficiaries Count',
    'BENES_WTH_PTAPTB_CNT': 'Part A and B Beneficiaries Count',
    'BENES_FFS_CNT': 'Fee-For-Service Beneficiaries Count',
    'BENES_MA_CNT': 'Medicare Advantage Beneficiaries Count',
    'MA_PRTCPTN_RATE': 'Medicare Advantage Participation Rate',
    'BENE_AVG_AGE': 'Average Beneficiary Age'
}, inplace=True)
```
