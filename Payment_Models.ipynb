```python
import pandas as pd
```

```python
df_payment_models = pd.read_csv('2014-2023 Medicare Fee-for-Service Geographic Variation Public Use File.csv')
```

```python
df_payment_models.columns
```

```python
display(df_payment_models)
```

```python
df_pm_cleaned = df_payment_models.replace('*', pd.NA)
```

```python
missing_values = df_pm_cleaned.isna().sum()
missing_values[missing_values > 0]
```

```python
display(df_pm_cleaned)
```

```python
columns_to_extract = [
    'YEAR', 'BENE_GEO_DESC', 'BENE_GEO_LVL',
    'TOT_MDCR_STDZD_PYMT_PC', 'TOT_MDCR_PYMT_PC', 
    'IP_MDCR_PYMT_PC','IP_MDCR_STDZD_PYMT_PC',
    'OP_MDCR_PYMT_PC', 'OP_MDCR_STDZD_PYMT_PC',
    'SNF_MDCR_PYMT_PC', 'SNF_MDCR_STDZD_PYMT_PC',
    'HH_MDCR_PYMT_PC', 'HH_MDCR_STDZD_PYMT_PC',
    'HOSPC_MDCR_PYMT_PC', 'HOSPC_MDCR_STDZD_PYMT_PC',
    'EM_MDCR_PYMT_PC', 'EM_MDCR_STDZD_PYMT_PC',
    'FQHC_RHC_MDCR_PYMT_PC', 'FQHC_RHC_MDCR_STDZD_PYMT_PC']
```

```python
df_pm_trends = df_pm_cleaned[columns_to_extract]
```

```python
display(df_pm_trends)
```

```python
df_pm_trends = df_pm_trends[
    ~df_pm_trends['BENE_GEO_DESC'].str.contains("ZZ|UNKNOWN", na=False)]
```

```python
display(df_pm_trends)
```

```python
missing_values = df_pm_trends.isna().sum()
missing_values[missing_values > 0]
```

```python
df_missing = df_pm_trends[df_pm_trends.isna().any(axis=1)]
```

```python
df_missing.shape
```

```python
df_missing.sample(10)
```

```python
df_pm_trends.dtypes
```

```python
payment_cols = df_pm_trends.columns.difference(['YEAR', 'BENE_GEO_DESC', 'BENE_GEO_LVL'])

for col in payment_cols:
    df_pm_trends[col] = pd.to_numeric(df_pm_trends[col], errors='coerce')
```

```python
df_pm_trends.dtypes
```

```python
display(df_pm_trends)
```

```python
df_pm_trends['BENE_GEO_LVL'] = df_pm_trends['BENE_GEO_LVL'].astype(str).str.strip().str.title()
```

```python
us_state_abbrev = {
    'AL': 'Alabama', 'AK': 'Alaska', 'AZ': 'Arizona', 'AR': 'Arkansas', 'CA': 'California',
    'CO': 'Colorado', 'CT': 'Connecticut', 'DE': 'Delaware', 'FL': 'Florida', 'GA': 'Georgia',
    'HI': 'Hawaii', 'ID': 'Idaho', 'IL': 'Illinois', 'IN': 'Indiana', 'IA': 'Iowa',
    'KS': 'Kansas', 'KY': 'Kentucky', 'LA': 'Louisiana', 'ME': 'Maine', 'MD': 'Maryland',
    'MA': 'Massachusetts', 'MI': 'Michigan', 'MN': 'Minnesota', 'MS': 'Mississippi', 'MO': 'Missouri',
    'MT': 'Montana', 'NE': 'Nebraska', 'NV': 'Nevada', 'NH': 'New Hampshire', 'NJ': 'New Jersey',
    'NM': 'New Mexico', 'NY': 'New York', 'NC': 'North Carolina', 'ND': 'North Dakota', 'OH': 'Ohio',
    'OK': 'Oklahoma', 'OR': 'Oregon', 'PA': 'Pennsylvania', 'RI': 'Rhode Island', 'SC': 'South Carolina',
    'SD': 'South Dakota', 'TN': 'Tennessee', 'TX': 'Texas', 'UT': 'Utah', 'VT': 'Vermont',
    'VA': 'Virginia', 'WA': 'Washington', 'WV': 'West Virginia', 'WI': 'Wisconsin', 'WY': 'Wyoming',
    'DC': 'District of Columbia'
}
```

```python
df_pm_trends.loc[df_pm_trends['BENE_GEO_LVL'] == 'State', 'BENE_GEO_DESC'] = \
    df_pm_trends.loc[df_pm_trends['BENE_GEO_LVL'] == 'State', 'BENE_GEO_DESC'].map(us_state_abbrev)
```

```python
display(df_pm_trends)
```

```python
df_pm_trends = df_pm_trends.rename(columns={
    'YEAR': 'Year',
    'BENE_GEO_DESC': 'Geo_Entity',
    'BENE_GEO_LVL': 'Geo_Level',

    'TOT_MDCR_PYMT_PC': 'Total Actual Per Capita',
    'TOT_MDCR_STDZD_PYMT_PC': 'Total Standardized Per Capita',

    'IP_MDCR_PYMT_PC': 'IP Actual Per Capita',
    'IP_MDCR_STDZD_PYMT_PC': 'IP Standardized Per Capita',

    'OP_MDCR_PYMT_PC': 'OP Actual Per Capita',
    'OP_MDCR_STDZD_PYMT_PC': 'OP Standardized Per Capita',

    'SNF_MDCR_PYMT_PC': 'SNF Actual Per Capita',
    'SNF_MDCR_STDZD_PYMT_PC': 'SNF Standardized Per Capita',

    'HH_MDCR_PYMT_PC': 'Home Health Actual Per Capita',
    'HH_MDCR_STDZD_PYMT_PC': 'Home Health Standardized Per Capita',

    'HOSPC_MDCR_PYMT_PC': 'Hospice Actual Per Capita',
    'HOSPC_MDCR_STDZD_PYMT_PC': 'Hospice Standardized Per Capita',

    'EM_MDCR_PYMT_PC': 'E&M Actual Per Capita',
    'EM_MDCR_STDZD_PYMT_PC': 'E&M Standardized Per Capita',

    'FQHC_RHC_MDCR_PYMT_PC': 'FQHC Actual Per Capita',
    'FQHC_RHC_MDCR_STDZD_PYMT_PC': 'FQHC Standardized Per Capita'
})
```

```python
display(df_pm_trends)
```

```python
df_pm_trends.to_csv("medicare_payment_models.csv", index=False)
```
